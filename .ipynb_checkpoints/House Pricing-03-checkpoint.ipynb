### 2. 数据清洗

为数据清洗方便，先将train数据集与test数据集合并，后续训练再分开

缺失值处理

```python
df_all = pd.concat((df_train, df_test), axis = 0)
```

```python
df_all.shape
```

```python
df_all.info()
```

```python
# 观察df_all中的缺失值的比例
missing = df_all.isnull().sum()
missing_count = missing[missing > 0].sort_values(ascending=False)
missing_rate = missing_count / len(df_all)
missing_data = pd.concat([missing_count,missing_rate], axis=1, keys=['count','ratio'])
missing_data
```

```python
# PoolQC相关，查看PoolQC为空值，但PoolArea不为0的值
print(df_all[(df_all['PoolQC'].isnull()) & (df_all['PoolArea']!=0)][['PoolQC','PoolArea']])
print(df_all[(df_all['PoolQC'].notnull()) & (df_all['PoolArea']==0)][['PoolQC','PoolArea']])
```

```python
df_all['PoolQC'] = df_all['PoolQC'].replace(df_all[(df_all['PoolQC'].isnull()) & (df_all['PoolArea']!=0)]['PoolQC'], 'Ex')
```

```python
# FireplaceQu相关，查看FireplaceQu为空值，但Fireplaces不为0的值
print(df_all[(df_train['FireplaceQu'].isnull()) & (df_all['Fireplaces'] != 0)][['FireplaceQu']])
```

```python
# PoolQC用None填充
df_all['PoolQC'].fillna('None', inplace = True)

# MiscFeature用None填充
df_all['MiscFeature'].fillna('None', inplace = True)

# Alley用None填充
df_all['Alley'].fillna('None', inplace = True)

# Fence用None填充
df_all['Fence'].fillna('None', inplace = True)

# FireplaceQu用None填充
df_all['FireplaceQu'].fillna('None', inplace = True)
```

```python
# Garage相关变量的值缺失是由没有车库造成的，用None替代
GarageX = ['GarageYrBlt','GarageType','GarageFinish','GarageQual','GarageCond']
for i in GarageX:
    df_all[i].fillna('None', inplace = True)
```

```python
# Bsmt相关变量的值缺失是由没有地下室造成的，用None替代
BsmtX = ['BsmtFinType2','BsmtFinType1','BsmtCond','BsmtExposure','BsmtQual']
for i in BsmtX:
    df_all[i].fillna('None', inplace = True)
```

```python
# MasVnrArea，装饰石材面积，缺失值用0替代
df_all['MasVnrArea'].fillna(0, inplace = True)
```

```python
# MasVnrType，装饰石材类型，缺失值用None替代
df_all['MasVnrType'].fillna('None', inplace = True)
```

```python
# LotFrontage，房子同街道之间的距离，缺失值较多，但与SalePrice相关系数大于0.3，保留变量，缺失值用中值替代
df_all['LotFrontage'] = df_all.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))
```

```python
# Electrical，电力系统，缺失值直接删除整行
df_all = df_all.drop(df_all.loc[df_all['Electrical'].isnull()].index)
```

```python
list1 = ['MasVnrType','MSZoning','Functional','Utilities','SaleType','KitchenQual',
         'Exterior2nd','Exterior1st','Functional']
for i in list1:
    df_all[i].fillna('None', inplace = True)
```

```python
list2 = ['MasVnrArea','BsmtFinSF2','BsmtUnfSF','BsmtFinSF1','TotalBsmtSF','GarageCars','GarageArea','BsmtFullBath','BsmtHalfBath']

for i in list2:
    df_all[i].fillna(0, inplace = True)
```

```python
df_all.info()
```

```python
df_all.isnull().sum()[df_all.isnull().sum() > 0]
```

缺失值处理完毕，接下来处理异常值

```python
# 删除异常点
df_all.drop(df_all[(df_all['GrLivArea']> 4000)&(df_all['SalePrice']< 400000)].index,inplace=True)
df_all.drop(df_all[(df_all['TotalBsmtSF']> 6000)].index,inplace=True)
```

异常点处理完毕，接下来处理数据类型

```python
# 将df_all的变量按照object类型和非object分类
ob_col = df_all.columns[df_all.dtypes == 'object']
num_col = df_all.columns[df_all.dtypes != 'object']
```

```python
len(ob_col), ob_col
```

```python
len(num_col), num_col
```

```python
# num_col中的'MSSubClass','OverallQual','OverallCond','YearBuilt', 'YearRemodAdd', 'MiscVal', 'MoSold', 'YrSold'为非连续变量,可以转化为object
df_all[['MSSubClass','OverallQual','OverallCond','YearBuilt', 'YearRemodAdd', 'MiscVal', 'MoSold', 'YrSold']] = df_all[['MSSubClass','OverallQual','OverallCond','YearBuilt', 'YearRemodAdd', 'MiscVal', 'MoSold', 'YrSold']].astype(str)
```

```python
ob_col = df_all.columns[df_all.dtypes == 'object']
len(ob_col)
```

```python
num_col = df_all.columns[df_all.dtypes != 'object']
len(num_col)
```

```python
# 观察数值型数据的分布
f = pd.melt(df_all, value_vars = num_col)
g = sns.FacetGrid(f, col = 'variable',  col_wrap = 5, sharex = False, sharey = False)
g = g.map(sns.distplot, 'value')
```

```python
# 'LotFrontage', 'LotArea','BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', 'GrLivArea'类似于正态分布,对其log处理
log_list = ['LotFrontage', 'LotArea', 'TotalBsmtSF', '1stFlrSF', 'GrLivArea']
for i in log_list:
    df_all[i] = np.log1p(df_all[i])
```

```python
# 针对其余数值型变量，也做标准化处理
num_col2 = num_col.drop(log_list)
num_col2 = num_col.drop('SalePrice')
num_col2_means = df_all.loc[:, num_col2].mean()
num_col2_std = df_all.loc[:, num_col2].std()
df_all.loc[:, num_col2] = (df_all.loc[:, num_col2] - num_col2_means) / num_col2_std
```

```python
# 处理序列型数据
order_col = ['ExterQual', 'ExterCond', 'GarageQual', 'GarageCond','PoolQC','FireplaceQu', 'KitchenQual', 'HeatingQC', 'BsmtQual','BsmtCond']
order_map = {'Ex': 5,'Gd': 4,'TA': 3,'Fa': 2,'Po': 1,'None': 0}

for i in order_col:
    df_all[i] = df_all[i].map(order_map)
```

```python
# 为其他类别属性贴标签
df_all['BsmtExposure'] = df_all['BsmtExposure'].map({'None':0, 'No':1, 'Mn':2, 'Av':3, 'Gd':4})    
df_all['BsmtFinType1'] = df_all['BsmtFinType1'].map({'None':0, 'Unf':1, 'LwQ':2,'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ':6})
df_all['BsmtFinType2'] = df_all['BsmtFinType2'].map({'None':0, 'Unf':1, 'LwQ':2,'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ':6})
df_all['Functional'] = df_all['Functional'].map({'None':0,'Sal':1, 'Sev':2, 'Maj2':3, 'Maj1':4, 'Mod':5, 'Min2':6, 'Min1':7, 'Typ':8})
df_all['GarageFinish'] = df_all['GarageFinish'].map({'None':0, 'Unf':1, 'RFn':2, 'Fin':3})
df_all['Fence'] = df_all['Fence'].map({'None':0, 'MnWw':1, 'GdWo':2, 'MnPrv':3, 'GdPrv':4})
```

```python
# 为一些类别做1/0处理
df_all['MasVnrType'] = df_all.MasVnrType.replace({'BrkCmn': 1,'BrkFace': 1,'CBlock': 1,'Stone': 1,'None': 0})
df_all['SaleCondition'] = df_all.SaleCondition.replace({'Abnorml': 1,'Alloca': 1,'AdjLand': 1,'Family': 1,'Normal': 0,'Partial': 0})
df_all['CentralAir'] = df_all.CentralAir.replace({'Y': 1,'N': 0})
df_all['PavedDrive'] = df_all.PavedDrive.replace({'Y': 1,'N': 0})
```

```python
# 对售卖月份按照季度做处理
df_all['MoSold'] = df_all['MoSold'].map({'1':1,'2':1,'3':1,'4':2,'5':2,'6':2,'7':3,'8':3,'9':3,'10':4,'11':4,'12':4})
```

```python
df_train_dummy.shape, df_test_dummy.shape
```

### 3. 建模

```python
x_train = df_train_n.values
y_train = np.log1p(df_train_dummy['SalePrice'])
x_test = df_test_n.values
```

```python
# 定义验证函数
def rmse_cv(model):
    rmse= np.sqrt(-cross_val_score(model, x_train, y_train, scoring='neg_mean_squared_error', cv = 10))
    return(rmse)
```

##### Ridge Regression

```python
from sklearn.model_selection import cross_val_score, GridSearchCV, KFold
from sklearn.linear_model import Ridge, RidgeCV
```

```python
clf1 = RidgeCV(np.logspace(-3, 2, 50))
clf1.fit(x_train, y_train)
ridge_pre = np.expm1(clf1.predict(x_test))
score1 = rmse_cv(clf1)
print("\nRidge score: {:.4f} ({:.4f})\n".format(score1.mean(), score1.std()))
```

##### Lasso Regression

```python
from sklearn.linear_model import LassoCV, LassoLarsCV
```

```python
clf2 = LassoCV(alphas = [1, 0.1, 0.001, 0.0005,0.0003,0.0002, 5e-4])
clf2.fit(x_train, y_train)
lasso_pre = np.expm1(clf2.predict(x_test))
score2 = rmse_cv(clf2)
print('\nLasso score: {:.4f} ({:.4f})\n'.format(score2.mean(), score2.std()))
```

##### ELASTIC NET

```python
from sklearn.linear_model import ElasticNet, ElasticNetCV
```

```python
clf3 = ElasticNetCV(l1_ratio=[.1, .5, .7, .9, .95, .99, 1],alphas=[0.001, 0.05, 0.1, 0.3, 1, 3, 5, 10, 15, 30, 50, 75])
clf3.fit(x_train, y_train)
elastic_pre = np.expm1(clf3.predict(x_test))
score3 = rmse_cv(clf3)
print('\nElastic Net: {:.4f} ({:.4f})\n'.format(score3.mean(), score3.std()))
```

### 5. 输出结果

```python
ridge_pre = np.expm1(clf1.predict(x_test))
lasso_pre = np.expm1(clf2.predict(x_test))
elastic_pre = np.expm1(clf3.predict(x_test))
```

```python
y_final_1 = ridge_pre
y_final_2 = lasso_pre
y_final_3 = elastic_pre
y_final_4 = 0.15 * ridge_pre + 0.6 * lasso_pre + 0.25 * elastic_pre
```
